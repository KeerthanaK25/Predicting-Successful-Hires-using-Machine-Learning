```python
Num_val = {'Yes':1, 'No':0}
data['Having_Kids'] = data["Having_Kids"].apply(lambda x: Num_val[x])
data['Martial_Status'] = data["Martial_Status"].apply(lambda x: Num_val[x])
data['Spouse_Working_Status'] = data["Spouse_Working_Status"].apply(lambda x: Num_val[x])
data['Position_Asked_Met'] = data["Position_Asked_Met"].apply(lambda x: Num_val[x])
data['WFH_Offered'] = data["WFH_Offered"].apply(lambda x: Num_val[x])
data['Having_Other_Offers'] = data["Having_Other_Offers"].apply(lambda x: Num_val[x])

data['Referrenced_Not'] = data["Referrenced_Not"].apply(lambda x: Num_val[x])
data['Break_Experience'] = data["Break_Experience"].apply(lambda x: Num_val[x])
data['Offer_Accepted_YN'] = data["Offer_Accepted_YN"].apply(lambda x: Num_val[x])


data.head()
```

```python
data = pd.get_dummies(data)
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

target = data['Offer_Accepted_YN']
features = data.drop('Offer_Accepted_YN', axis = 1)



#create the train/test split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=10)

#Create the model and train
model = RandomForestClassifier()
model.fit(X_train,y_train)

#predict the results for test
test_pred = model.predict(X_test)

#test the accuracy
accuracy_score(y_test, test_pred)
```

```python
test_pred
```

```python
feat_importances = pd.Series(model.feature_importances_, index=features.columns)
feat_importances = feat_importances.nlargest(10)
feat_importances.plot(x='PERCENTAGE',y = 'FACTORS',kind='barh',ax = None)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=features.columns)
feat_importances = feat_importances.nsmallest(10)
feat_importances.plot(kind='barh')
```
